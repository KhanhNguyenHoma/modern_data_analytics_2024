```python
import pandas as pd
import numpy as np
```

```python

# Load the intervention coordinates
intervention_coordinates = pd.read_csv('../transformed_data/intervention_coordinates.csv')

# Load the potential AED locations
potential_aed_locations = pd.read_csv('../transformed_data/potential_aed_locations.csv')
```

```python
from sklearn.metrics.pairwise import haversine_distances
from math import radians

# Convert latitude and longitude to radians and store in numpy arrays
intervention_coordinates_rad = np.radians(intervention_coordinates[['lat', 'lon']])
potential_aed_locations_rad = np.radians(potential_aed_locations[['lat', 'lon']])

# Compute haversine distances
distances = haversine_distances(intervention_coordinates_rad, potential_aed_locations_rad)

# Convert to kilometers
distances *= 6371000/1000  # multiply by Earth radius to get kilometers

# Create a DataFrame for the distances
distances_df = pd.DataFrame(distances, index=intervention_coordinates['id'], columns=potential_aed_locations.index)

distances_df
```
